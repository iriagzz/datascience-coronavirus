Após a análise dos dados globais, notou-se que os países Nórdicos não se comportavam de maneiras "harmoniosa". A Suécia surge como um outlier.
Esta análise teve como objetivo verificar os dados globais, em seguida verificar o comportamento dos países. Os dados foram obtidos no portal Our World in Data (https://ourworldindata.org/coronavirus), em 09 de junho de 2020.

# Análise dos Países Nórdicos

```python
df = pd.read_csv("owid-covid-data09-06.csv", index_col=0, header=0)
```

```python
df.dropna()
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df_nordicos = df.query('location in ["Sweden", "Norway", "Finland", "Denmark", "Iceland"]')
```

```python
df_nordicos.head()
```

# Total Mortes

```python
date=df_nordicos.loc["DNK", ["date"]]
dnk=df_nordicos.loc["DNK", ["total_deaths"]]
swe=df_nordicos.loc["SWE", ["total_deaths"]]
nor=df_nordicos.loc["NOR", ["total_deaths"]]
isl=df_nordicos.loc["ISL", ["total_deaths"]]
fin=df_nordicos.loc["FIN", ["total_deaths"]]
```

```python
plt.plot_date(date, dnk, label="Dinamarca", linestyle='solid', marker='None', linewidth=3.0)
plt.plot_date(date, swe, label="Suécia", linestyle='solid', marker='None', linewidth=3.0)
plt.plot_date(date, nor, label="Noruega", linestyle='solid', marker='None', linewidth=3.0)
plt.plot_date(date, isl, label="Islândia", linestyle='solid', marker='None', linewidth=3.0)
plt.plot_date(date, fin, label="Finlândia", linestyle='solid', marker='None', linewidth=3.0)

plt.gcf().autofmt_xdate()
date_format = mpl_dates.DateFormatter('%d-%m-%Y')
plt.gca().xaxis.set_major_formatter(date_format)
plt.legend()

plt.ylabel('Total de Mortes')
```

# Total Casos

```python
date=df_nordicos.loc["DNK", ["date"]]
dnk1=df_nordicos.loc["DNK", ["total_cases"]]
swe1=df_nordicos.loc["SWE", ["total_cases"]]
nor1=df_nordicos.loc["NOR", ["total_cases"]]
isl1=df_nordicos.loc["ISL", ["total_cases"]]
fin1=df_nordicos.loc["FIN", ["total_cases"]]
```

```python
plt.plot_date(date, dnk1, label="Dinamarca", linestyle='solid', marker='None', linewidth=3.0)
plt.plot_date(date, swe1, label="Suécia", linestyle='solid', marker='None', linewidth=3.0)
plt.plot_date(date, nor1, label="Noruega", linestyle='solid', marker='None', linewidth=3.0)
plt.plot_date(date, isl1, label="Islândia", linestyle='solid', marker='None', linewidth=3.0)
plt.plot_date(date, fin1, label="Finlândia", linestyle='solid', marker='None', linewidth=3.0)

plt.gcf().autofmt_xdate()
date_format = mpl_dates.DateFormatter('%d-%m-%Y')
plt.gca().xaxis.set_major_formatter(date_format)
plt.legend()

plt.ylabel('Total de Casos')
```

# Índice de Rigidez

```python
dnk2=df_nordicos.loc["DNK", ["stringency_index"]]
swe2=df_nordicos.loc["SWE", ["stringency_index"]]
nor2=df_nordicos.loc["NOR", ["stringency_index"]]
isl2=df_nordicos.loc["ISL", ["stringency_index"]]
fin2=df_nordicos.loc["FIN", ["stringency_index"]]
```

```python
plt.plot_date(date, dnk2, label="Dinamarca", linestyle='solid', marker='None', linewidth=2.0)
plt.plot_date(date, swe2, label="Suécia", linestyle='solid', marker='None', linewidth=2.0)
plt.plot_date(date, nor2, label="Noruega", linestyle='solid', marker='None', linewidth=2.0)
plt.plot_date(date, isl2, label="Islândia", linestyle='solid', marker='None', linewidth=2.0)
plt.plot_date(date, fin2, label="Finlândia", linestyle='solid', marker='None', linewidth=2.0)

plt.gcf().autofmt_xdate()
date_format = mpl_dates.DateFormatter('%d-%m-%Y')
plt.gca().xaxis.set_major_formatter(date_format)
plt.legend()

plt.ylabel('Índice de Rigidez')
```

# Novos Casos

```python
start_date = '2020-3-1'
end_date = '2020-6-9'
mask = (df_nordicos['date'] > start_date) & (df_nordicos['date'] <= end_date)
df_nordicos2=df_nordicos.loc[mask]
date2=df_nordicos2.loc["DNK", ["date"]]
dnk3=df_nordicos2.loc["DNK", ["new_cases"]]
swe3=df_nordicos2.loc["SWE", ["new_cases"]]
nor3=df_nordicos2.loc["NOR", ["new_cases"]]
isl3=df_nordicos2.loc["ISL", ["new_cases"]]
fin3=df_nordicos2.loc["FIN", ["new_cases"]]
```

```python
plt.plot_date(date2, dnk3, label="Dinamarca", linestyle='solid', marker='None', linewidth=2.0)
plt.plot_date(date2, swe3, label="Suécia", linestyle='solid', marker='None', linewidth=2.0)
plt.plot_date(date2, nor3, label="Noruega", linestyle='solid', marker='None', linewidth=2.0)
plt.plot_date(date2, isl3, label="Islândia", linestyle='solid', marker='None', linewidth=2.0)
plt.plot_date(date2, fin3, label="Finlândia", linestyle='solid', marker='None', linewidth=2.0)

plt.gcf().autofmt_xdate()
date_format = mpl_dates.DateFormatter('%d-%m-%Y')
plt.gca().xaxis.set_major_formatter(date_format)
plt.legend()

plt.ylabel('Novo Casos')
```

# correlação com máximos e stringency médio

```python
df_corr = df_nordicos.groupby('location').agg({'total_deaths': np.max, 'total_cases': np.max, 'stringency_index' : np.mean})
```

```python
df_corr.dropna()
```

```python
df_corr = df_corr.corr()
```

```python
df_corr
```

Esta análise fica comprometida pela presença de outliers, o que tende a distorcer o valor da média e, por conseqüência, o valor do coeficiente de correlação. Dessa forma, a presença de outliers pode comprometer fortemente as estimativas dos pesquisadores.
